Visualitation
- Line plots - growth / delay
- Bar Charts - individual comparision b/w values
- Pie Charts - individual comparision b/w values by percentage
- scatter plots - comparing 2 or 3 values by their grouping/clusters
- histograms - frequency wise distribution of data
- area plot - same as line
- 3rd plot - for 3 dimensional data understanding

```python
t = np.arange(-np.pi*3, np.pi*3, 0.5)
x = np.sin(t)
y = np.cos(t)
```

```python
plt.plot(t,x,'ro:')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.plot(t,y,'gs--')
plt.plot(t,x,'r--')
plt.show()
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
data.sort_values(by='age', inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.plot(data['age'],'ro')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.bar(x=data.index,height=data['age'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
data['fare'].tail(25).plot(kind='barh', ax=ax, color=['red','purple','black'])
plt.show()
```

```python
data2=sns.load_dataset('penguins')
data2.head()
```

```python
data2['species'].value_counts()
```

```python
data2['species'].value_counts().plot(kind='bar',color=['orange','skyblue','brown'])
```

```python
fig,ax = plt.subplots(figsize=(5,5))
vals=data2['species'].value_counts().plot(kind='bar',color=['orange','skyblue','brown'])
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(5,5))
vals=data2['species'].value_counts().plot(kind='pie',colors=['orange','skyblue','brown'])
plt.show()
```

```python
fig, ax = plt.subplots(facecolor='pink')
data2['species'].value_counts().plot(kind='pie',
                                    colors=['orange','skyblue','brown'],
                                    autopct='%.1f',
                                    shadow=True)
plt.show()
```

```python
px.scatter_3d(data2, x='bill_length_mm', y='bill_depth_mm', z='flipper_length_mm',
            color='species', symbol='island')
```
